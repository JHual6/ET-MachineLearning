### Identificacin y eliminacin de Outliers

```python
def eliminar_outliers(df):
    df_sin_outliers = df.copy()
    columnas_numericas = df.select_dtypes(include=[np.number]).columns

    for col in columnas_numericas:
        Q1 = df_sin_outliers[col].quantile(0.25)
        Q3 = df_sin_outliers[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df_sin_outliers = df_sin_outliers[(df_sin_outliers[col] >= lower_bound) & (df_sin_outliers[col] <= upper_bound)]

    return df_sin_outliers

df_sin_outliers = eliminar_outliers(df_sin_nulos)
print(f"Cantidad de registros antes de eliminar outliers: {len(df_sin_nulos)}")
print(f"Cantidad de registros despus de eliminar outliers: {len(df_sin_outliers)}")

plt.figure(figsize=(12, 6))
df_sin_outliers.boxplot()
plt.title("Datos despus de eliminar los outliers")
plt.show()
```

La cantidad de Outliers es muy alta, lo cual puede afectar negativamente a los futuros modelos que se implementarn, aunque sean muchos datos los que se eliminarn con los outlinders el rendimiento de los modelos aumentar considerablemente, los datos iniciales son 42.373, pero con la eliminacin de los Outliers pasan a ser 8.996, lo cual reduce el conjunto de datos en un 78.77% en comparacin con el inicial, lo cual es ms de un tercio, pero garantizar que los datos sean ms precisos.

```python
k_values = [4, 5, 6, 7, 8]
results = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10, max_iter=300)
    kmeans_labels = kmeans.fit_predict(modelos_no_supervisados_scaled)

    silhouette = silhouette_score(modelos_no_supervisados_scaled, kmeans_labels)
    calinski = calinski_harabasz_score(modelos_no_supervisados_scaled, kmeans_labels)
    davies_bouldin = davies_bouldin_score(modelos_no_supervisados_scaled, kmeans_labels)

    results.append({
        'k': k,
        'Silhouette Score': silhouette,
        'Calinski-Harabasz Score': calinski,
        'Davies-Bouldin Score': davies_bouldin
    })

results_df = pd.DataFrame(results)
print(results_df)

best_k = results_df.loc[
    (results_df['Silhouette Score'] == results_df['Silhouette Score'].max()) |
    (results_df['Calinski-Harabasz Score'] == results_df['Calinski-Harabasz Score'].max()) |
    (results_df['Davies-Bouldin Score'] == results_df['Davies-Bouldin Score'].min())
]

if best_k.empty:
    best_k = results_df.loc[
        (results_df['Silhouette Score'] == results_df['Silhouette Score'].max()) &
        (results_df['Davies-Bouldin Score'] == results_df['Davies-Bouldin Score'].min())
    ]

best_k_value = best_k['k'].values[0]
print(f"Mejor valor de k: {best_k_value}")
```
